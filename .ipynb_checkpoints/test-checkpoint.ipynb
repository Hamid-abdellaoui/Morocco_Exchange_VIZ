```python

millnames = ['',' Mille',' Million',' Milliard',' Mille milliards']

def millify(n):
    n = float(n)
    millidx = max(0,min(len(millnames)-1,
                        int(math.floor(0 if n == 0 else math.log10(abs(n))/3))))

    return '{:.0f}{}'.format(n / 10**(3 * millidx), millnames[millidx])
```

```python
### importing Data and extracting 
Export_total = pd.read_csv('data/Export_total.csv')
Export_total.head(1)
```

```python
new_df = pd.read_csv('data/new_df.csv')
new_df.head()
```

```python
flux=new_df.groupby(['Libellé du pays','Libellé du flux']).sum()
flux=flux.transpose()
flux.set_index(pd.date_range(start='2018-01-01', periods=42,freq='M'),inplace=True)
flux.head()
```
